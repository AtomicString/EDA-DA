```python
time_vals = gunnels['Time']
time_bins = pd.cut(time_vals, [600, 700, 800, 900, 1000])
time_bins
```

```python
time_bins.value_counts()
```

```python
time_vals.quantile([0.25, 0.5, 0.75])
```

```python
IQR = np.percentile(time_vals, 75) - np.percentile(time_vals, 25)
IQR
val1 = np.percentile(time_vals, 25) - 1.5 * IQR
val2 = np.percentile(time_vals, 75) + 1.5 * IQR

time_vals[~time_vals.between(val1, val2)] # Outliers
```

```python
non_outlier_time_vals = time_vals[time_vals.between(val1,val2)]
non_outlier_time_vals #Without outliers
```
